```python
from qiskit_ionq import IonQProvider
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, transpile

# 1. Configuración de registros (El "Hardware" virtual)
# Definimos nombres claros para no perdernos entre índices
qubits = QuantumRegister(2, name="q_ion")
bits = ClassicalRegister(2, name="c_out")
qc = QuantumCircuit(qubits, bits)

# 2. Construcción del Circuito paso a paso
# Usamos los nombres de los registros en lugar de solo números
qc.h(qubits[0])            # Superposición en el primer qubit
qc.cx(qubits[0], qubits[1]) # Entrelazamos el 0 con el 1

# 3. Medición Explícita
# Aquí ves exactamente qué qubit va a qué slot de memoria clásica
qc.measure(qubits[0], bits[0])
qc.measure(qubits[1], bits[1])

# 4. Conexión con IonQ
provider = IonQProvider('Wtl76IgoWaLTGjpv9R3GANMF92KVmXA9')
backend = provider.backends()[0]

# 5. Transpilación y Ejecución
qc_ionq = transpile(qc, backend=backend, optimization_level=1)
print(f"Ejecutando {qc.name} en {backend.name}...")

job = backend.run(qc_ionq, shots=100)
result = job.result()

print("Resultados detallados:", result.get_counts())
```
